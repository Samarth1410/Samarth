```python
import pandas as pd
from google.colab import files
```

```python
sales_data = pd.read_csv("Sales.csv")
```

```python
sales_data.columns
```

```python
sales_data.head(4)
```

```python
trim_sales = sales_data[['Region', 'Country', 'Item Type', 'Total Revenue', 'Total Cost', 'Total Profit']]
```

```python
trim_sales
```

```python
trim_sales.describe()
```

```python
regionwise_sales =trim_sales.groupby('Region').sum()
```

```python
regionwise_sales
```

```python
regionwise_sales.plot.bar(stacked = True) #Plotting using pandas
```

```python
country_sales =trim_sales.groupby('Country').sum()
```

```python
country_sales
```

```python
sort_country_sales = country_sales.sort_values(by = 'Total Profit', ascending = False)
```

```python
sort_country_sales.head().plot.bar()
```
