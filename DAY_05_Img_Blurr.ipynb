```python
#AIM : To blurr the phone screen
```

```python
plt.imshow(orig_img[175:225,350:415])
```

```python
img_portion = orig_img[175:225,350:415]
```

```python
blurr_part = cv.blur(img_portion , (10,10))
```

```python
plt.imshow(blurr_part)
```

```python
#Blur part done!.. now write back to orig img
```

```python
orig_img[175:225,350:415] = blurr_part
```

```python
plt.imshow(orig_img)
```

```python
cv.imwrite('blurr_img.png' , orig_img)
```
