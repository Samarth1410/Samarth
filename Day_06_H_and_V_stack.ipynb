```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
from google.colab import files
```

```python
samples = np.linspace(1,7,200)
plt.plot(np.sin(samples) , label = 'sin')
plt.plot(np.cos(samples) ,label = 'cos')
plt.legend()
plt.savefig('sincos.png')
```

```python
raw_img = cv.imread('Dragon_Ball.jpg' , 1)
```

```python
raw_img.shape #3 represents no. of channels. i.e: Red, Blue, Green
```

```python
plt.imshow(raw_img)
```

```python
vertically_stack_image = np.vstack((raw_img,raw_img))
```

```python
plt.imshow(vertically_stack_image)
```

```python
horizontally_stack_image = np.hstack((vertically_stack_image,vertically_stack_image))
plt.imshow(horizontally_stack_image)
```

```python
cv.imwrite('v + h stack image.png' , horizontally_stack_image)
```
